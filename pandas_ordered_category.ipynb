```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/staff.csv', index_col='empID', 
                 thousands=',')
df
```

```python
df.info()
```

```python
df.sort_values('position')
```

```python
df.position.unique()
```

```python
df['position_cat']=pd.Categorical(df['position'], 
            categories=['', '', '', '', ''], 
            ordered=True)
df
```

```python
df.position_cat.cat.categories
```

```python
pd.Series(df.position_cat.cat.categories)
```

```python
list(enumerate(df.position_cat.cat.categories))
```

```python
df.position_cat
```

```python
df.position_cat.cat.codes
```

```python
df['position_codes']=df.position_cat.cat.codes
df
```

```python
df.sort_values('position_cat')
```

```python
df.pivot_table(index='position', columns='sex', values='fname', aggfunc='count', fill_value=0).astype(int)
```

```python
df.pivot_table(index='position_cat', columns='sex', values='fname', aggfunc='count', fill_value=0).astype(int)
```

```python
pd.crosstab(df.sex, df.position)
```

```python
pd.crosstab(df.sex, df.position_cat)
```
