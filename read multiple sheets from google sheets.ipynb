```python
import pandas as pd

print(f'pandas   version = {pd.__version__}')
print(pd.Timestamp.now())
```

## pandas a specific sheet
* Google sheets: https://docs.google.com/spreadsheets/d/15PUfhKBAGs-_z3sGaH10AwyjYEKHLQlJUlrAEDK6jVw/edit#gid=0
* enable sharing mode in Google Sheets
* copy Google Sheets URL
* pass sheet_name = # if sheet_name = None, read all sheets into dict {key: df}

```python
# https://docs.google.com/spreadsheets/d/15PUfhKBAGs-_z3sGaH10AwyjYEKHLQlJUlrAEDK6jVw/edit?usp=sharing
url='https://docs.google.com/spreadsheets/d/15PUfhKBAGs-_z3sGaH10AwyjYEKHLQlJUlrAEDK6jVw/export?format=xlsx'
df=pd.read_excel(url, sheet_name=1)
df
```

## read multiple sheets by passing sheet index as list

```python
dfs=pd.read_excel(url, sheet_name=list(range(1, 4)))
print(len(dfs))
print(dfs.keys())
```

```python
dfs=pd.read_excel(url, sheet_name=[0, 2, 4, 5])
print(len(dfs))
print(dfs.keys())
```

```python
dfs[0]
```

## read multiple sheets by passing list of sheet names

```python
dfs=pd.read_excel(url, sheet_name=['CRIME', 'COMEDY', 'HORROR'])
print(len(dfs))
print(dfs.keys())
```

```python
dfs['COMEDY']
```

## pandas read all sheets
* pass sheet_name = None will read all sheets into dict {key: df}

```python
dfs=pd.read_excel(url, sheet_name=None)
```

```python
print(len(dfs))
print(dfs.keys())
```

```python
dfs['COMEDY']
```

## concat all sheets

```python
dm=pd.concat(dfs)
dm
```

```python
dfs.keys()
```

## concat and reset_index

```python
dm=pd.concat(dfs, keys=dfs.keys(), names=['Genre', 'no'])
d_all=dm.reset_index().drop(columns=['no'])
d_all
```

## save to Excel

```python
d_all.to_excel('all_movies.xlsx', index=False)
```
