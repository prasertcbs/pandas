```python
def bkkapi(resource_id):
	url=f"https://data.bangkok.go.th/api/3/action/datastore_search?resource_id={resource_id}"
	r=requests.get(url)
	j=r.json()
	# j['result']['records']
	df=pd.DataFrame(j['result']['records'])
	return df

# https://data.bangkok.go.th/dataset/65eb2ee6-efc6-4045-8e3e-ef451b844953/resource/95bd1379-177e-4dd8-9374-c0ce307bede9/download/-..-64.csv
dt=bkkapi("95bd1379-177e-4dd8-9374-c0ce307bede9")
dt
```

```python
def download(url):
	if (url[-4:] == ".xls") or (url[-5:] == ".xlsx"):
		df = pd.read_excel(url)
	else:
		df = pd.read_csv(url)
	return df

# url = "https://data.bangkok.go.th/dataset/c051d4a7-8418-49c1-86b0-5a5ea8511c33/resource/5cd749fe-11a8-457d-b88f-2151ac33fe27/download/fin_600ea1e3c81a7.csv"
url = "https://data.bangkok.go.th/dataset/e53ef05f-32b2-4b06-b800-cad2702f426e/resource/272b8fbc-6146-40d6-be7c-335ea30c2f7e/download/wq-2-reused_water-2561-2563.xlsx"
df=download(url)
df[:5]
```
